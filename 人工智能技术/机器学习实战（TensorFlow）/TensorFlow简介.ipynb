## 初学机器学习
机器学习的范例是我们有数据，这些数据是被标注的，并且我们想搞清楚匹配数据与标签的规则。如：

$X = -1, 0, 1, 2, 3, 4$

$Y = -3, -1, 1, 3, 5, 7$

我们通过大脑思考依次尝试后后发现它们之间的关系是 $Y=2X-1$。这和机器学习的过程非常相似。下面编写 TensorFlow 代码用一个神经网络解决这个问题，采用 TensorFlow Keras API。

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

re = Dense(units=1, input_shape=[1])
model = Sequential([re]) 
model.compile(optimizer='sgd', loss='mean_squared_error')

xs = np.array([-1,0,1,2,3,4], dtype=int)
ys = np.array([-3,-1,1,3,5,7], dtype=int)

model.fit(xs, ys, epochs=500)

print(model.predict([10]))
```

```python
print("Here is what I learned : {}".format(re.get_weights()))
```

使用 TensorFlow 时，使用 Sequential 来定义层。Sequential 可以指定每一层看起来是什么样的。下面的代码 Sequential 中只有一行，所以这是一个一层神经网络。
